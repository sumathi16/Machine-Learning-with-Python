NOT SPECIFIED  -0<br>
LOW  - 1<br>
MEDIUM - 2<br>
HIGH -3<br>
CRITICAL -4<br>

Define a function 
 - input is a string
 - return value must be 0,1,2,3 or 4

```python
def cat_to_num(x):
    if x=='NOT SPECIFIED':
        return 0
    elif x == 'LOW':
        return 1
    elif x=='MEDIUM':
        return 2
    elif x == 'HIGH':
        return 3
    elif x=='CRITICAL':
        return 4
```

```python
data['Order_Priority_Number'] =\
        data['Order_Priority'].apply(cat_to_num)
data['Order_Priority_Number'].value_counts()
```
